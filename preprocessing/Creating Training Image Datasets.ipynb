```python
# loading the csv with all instances
# csv_filename = 'data/train_cv_224/training_labels_handOverFace_unique.csv'
csv_filename = 'data/train_cv_224/all_net_images.csv'
imagenames = pd.read_csv(csv_filename, sep=' ')
imagenames.groupby('label').count()
```

```python
# remove the 'not sitting' category
# imagenames = imagenames[(imagenames.label != 0) & (imagenames.label != 2)]
imagenames = imagenames[(imagenames.label != 0)]
imagenames.groupby('label').count()
```

```python
# changing labels values
# imagenames.replace({'label': {1: 0, 3: 1, 4: 2, 5: 3, 6: 4}}, inplace=True)
imagenames.replace({'label': {1:0, 2:1, 3:2, 4:3, 5:4}}, inplace=True)
imagenames.groupby('label').count()
```

```python
# randomly removing instances to make the same size for each category
g = imagenames.groupby('label')
imagenames1 = g.apply(lambda x: x.sample(g.size().min()).reset_index(drop=True))
imagenames1 = imagenames1.droplevel('label')
imagenames1.groupby('label').count()
# imagenames1.groupby(imagenames1['label']).count() 
```

```python
# creating dummy variables for the cats
label_dummies = pd.get_dummies(imagenames1.label)
imagenames1 = pd.concat([imagenames1, label_dummies], axis=1)
del imagenames1['label']
imagenames1
```

```python
imagenames1.to_csv('data/train_cv_224/net_handOverFace_n447_5cats.csv', index=False)
```
